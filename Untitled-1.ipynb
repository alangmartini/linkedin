```python
import inputs

class JobInfo:
    def __init__(self, driver):
        # Get the element with the class job-view-layout jobs-details
        self.job_holder = driver.find_element(By.CLASS_NAME, 'jobs-details')
        self.job_description_holder = self.job_holder.find_element(By.CLASS_NAME, 'jobs-description__container')

    def get_job_title(self):
        return self.job_holder.find_element(By.TAG_NAME, 'a').text

    def get_job_description(self):
        return self.job_description_holder.text

    def get_easy_apply_button(self):
        return driver.find_element(By.XPATH, r"//span[@class='artdeco-button__text' and text()='Easy Apply']")

      

class ApplicationManager():
    IN_PROGRESS = 0
    SECOND_FROM_THE_LAST = 1
    LAST = 2
    
    def __init__(self, driver):
        self.driver = driver
        self.application_panel = self.driver.find_element(
            By.CSS_SELECTOR,
            'div.artdeco-modal.jobs-easy-apply-modal[data-test-modal][role="dialog"]'
        )
        self.is_done = False
        
        
    def check_is_final(self):
        # get the with using XPATH where the text is Submit application
        try:
            element = self.get_submit_button()
            
            if element.text == "Submit application":
                self.is_done = True
                return True
            else:
                return False
        except:
            return False

    def get_next_button(self):
        return self.application_panel.find_element(
            By.XPATH,
            r"//button[@aria-label='Continue to next step']"
        )

    def click_next_button(self):
        button = self.get_next_button()
        button.click()
        
    def get_review_button(self):
        return self.application_panel.find_element(
            By.XPATH,
            r"//button[@aria-label='Review your application']"
        )

    def click_review_button(self):
        button = self.get_review_button()
        button.click()
        
    def get_submit_button(self):
        return self.application_panel.find_element(
            By.XPATH,
            r"//button[@aria-label='Submit application']"
        )

    def click_submit_button(self):
        button = self.get_submit_button()
        button.click()
        self.is_done = True

    def get_is_done(self):
        return self.is_done

    def get_current_status(self):
        """ 
            A primeira página:

            Pode ser next e pedirá email, phone e mobile.
            Pode ja permitir dar submit, nesse caso será 'Submit application'.
            
            Se existir um botão 'Next', entamos em andamento.
            Se existir um botão 'Review', estamos na PENULTIMA etapa.
            O ultimo botão e que marca o final é 'Submit application'

            0 -> Em andamento
            1 -> Em finalização
            2 -> Ultima etapa
        """
        try:
           self.get_next_button()
           return self.IN_PROGRESS
        except:
            try:
                self.get_review_button()
                return self.SECOND_FROM_THE_LAST
            except:
                self.get_submit_button()
                return self.LAST


    def parse_inputs(self):
        radios = []
        texts = []
        selects = []

        options = self.driver.find_elements(
            By.CSS_SELECTOR,
            ".jobs-easy-apply-form-section__grouping"
        )

        for option in options:
            if inputs.RadioLinkedinInput(option).check_is_radio_type():
                radios.append(inputs.RadioLinkedinInput(option))
            elif inputs.TextLinkedinInput(option).check_is_text_type():
                texts.append(inputs.TextLinkedinInput(option))
            elif inputs.SelectLinkedinInput(option).check_is_select_type():
                selects.append(inputs.SelectLinkedinInput(option))
            else:
                print("Not found")

        return [radios, texts, selects]
    
    def fill_inputs(self):
        radios, texts, selects = self.parse_inputs()

        for radio in radios:
            radio.fill_input()
            time.sleep(0.5)
        for text in texts:
            text.fill_input()
            time.sleep(0.5)
        for select in selects:
            select.fill_input()
            time.sleep(0.5)

    @staticmethod
    def click_continue_applying_button(driver):
        button = driver.find_element(By.XPATH, r"//span[text()='Continue applying']")
        button.click()
        
    def click_dismiss_button(self):
        button = driver.find_element(By.XPATH, "//button[@aria-label='Dismiss']")
        button.click()
        
    def fill_applitcation(self):
        pass
    
```

```python
class JobListManager():
    def __init__(self, driver):
        self.get_job_list = driver.find_element(By.CSS_SELECTOR, "#main > div > div.scaffold-layout__list-detail-inner.scaffold-layout__list-detail-inner--grow > div.scaffold-layout__list > div > ul")

    def get_jobs_list(self):
        return self.get_job_list.find_elements(By.XPATH, './*')
    
    def click_on_job_by_index(self, index):
        item = self.get_jobs_list()[index]
        href = item.find_element(By.TAG_NAME, 'a')
        href.click()

    def click_on_job(self, job):
        href = job.find_element(By.TAG_NAME, 'a')
        href.click()

    def click_on_next_page(self, driver, page):
        button = driver.find_element(By.XPATH, f"//button[@aria-label='Page {page}']")
        button.click()
        
        

job_list_man = JobListManager(driver)
for a in range(8):
    for i, job in enumerate(job_list_man.get_jobs_list()):
        job_list_man.click_on_job_by_index(i)
        
        time.sleep(0.3)
        
        job_data = JobInfo(driver)
        
        # If there is no easy apply button, already applied
        try:
            job_data.get_easy_apply_button().click()
        except:
            time.sleep(0.3)
            continue
        
        # Sometimes it shows an alert of safety
        # we can just dismiss it
        try:
            ApplicationManager.click_continue_applying_button(driver)
        except:
            pass
        
        time.sleep(1)

        app_man = ApplicationManager(driver)
        
        if app_man.get_current_status() == app_man.LAST:
            app_man.click_submit_button()
            time.sleep(1)
            app_man.click_dismiss_button()
            continue

        app_man.click_next_button()
        time.sleep(0.5)
        app_man.click_next_button()

        current_status = app_man.get_current_status()
        
        while current_status == 0:
            
            app_man.fill_inputs()
            current_status = app_man.get_current_status()
            time.sleep(1)

        time.sleep(1)
        
        app_man.fill_inputs()
        
        # Here, we are on review submission
        app_man.click_review_button()

        time.sleep(0.5)

        app_man.click_submit_button()

        time.sleep(3)
        app_man.click_dismiss_button()
        
    job_list_man.click_on_next_page(driver, a+2)

print('done')
        
        
        
    
```
